```python
# Load libraries.....

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier                    # import Decision Tree Classifier.
from sklearn.model_selection import train_test_split               # import train_test_split function.
from sklearn import metrics                                        # import scikit-learn metrics for accuracy calculation.
from sklearn.metrics import confusion_matrix
```

```python
import matplotlib as plt
import seaborn as sns
%matplotlib inline 
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# load dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data.head(10)
```

```python
Y = iris.target
print('y:',Y)
```

```python
# Split dataset into train set and test set

X_train, X_test, y_train, y_test = train_test_split(data,Y, test_size=0.40, random_state=52)   
```

```python
# Creating Decision Tree Classifier

fin = DecisionTreeClassifier()
fin.fit(X_train,y_train)
```

```python
y_predict = fin.predict(X_test)
print(y_predict)
```

```python
# accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_predict))
```

```python
feature_n=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
class_n=['Iris-setosa','Iris-versicolor','Iris-virginica']

fig, axes = plt.subplots(figsize = (4,4), dpi = 400)
tree.plot_tree(fin, feature_names = feature_n, class_names =class_n , filled = False);
```
